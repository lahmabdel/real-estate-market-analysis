```python
bas_appart=pd.DataFrame(appart.groupby(['cat_commune'])["nom_commune"].unique()).iloc[0,0]
moyen_appart=pd.DataFrame(appart.groupby(['cat_commune'])["nom_commune"].unique()).iloc[1,0]
haut_appart=pd.DataFrame(appart.groupby(['cat_commune'])["nom_commune"].unique()).iloc[2,0]
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax = sns.boxplot(x="cat_commune", y="valeur_m2", data=avg_cat)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols("valeur_m2 ~ cat_commune", data= avg_cat).fit()
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```

```python
import scipy.stats as stats
stats.shapiro(model.resid)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize= (10, 10))
ax = fig.add_subplot(111)

normality_plot, stat = stats.probplot(model.resid, plot= plt, rvalue= True)
ax.set_title("Probability plot of model residual's", fontsize= 20)
ax.set

plt.show()
```

```python
'''ANOVA results cannot be interpreted, because of Shapiro Wilik Test and the QQ-Plot which shows that residuals is heavy-tailed
Instead, Let use a non-parametric Test based on ranks, Kruskal Wallis similar to ANOVA use to compare medians, before doing this, variables used in the test should be discret.
Kruksal Hypothesis :
- H0 : medians of all groups are equal
- H1 : at least one population median of one group is different from the population median of at least one other group.'''
```

```python
from scipy import stats
stats.kruskal(avg_cat.cat_commune, avg_cat['valeur_m2'])
```
