```python
airbnb=pd.read_csv('airbnb_data_simple.csv',index_col='Unnamed: 0')
vente=pd.read_csv('house_selling_data.csv',index_col='Unnamed: 0')
#location=
```

```python
airbnb['code_departement']=airbnb['code_departement'].astype(str)
vente['code_departement']=vente['code_departement'].astype(str)
```

```python
airbnb['nom_commune']=airbnb['code_departement']+airbnb['nom_commune']
vente['nom_commune']=vente['code_departement']+vente['nom_commune']
```

```python
appart=pd.read_csv('avg_m2_appart.csv',index_col='Unnamed: 0')
appart.groupby(['cat_commune'])["nom_commune"].unique()
```

```python
moyen_appart=pd.DataFrame(appart.groupby(['cat_commune'])["nom_commune"].unique()).iloc[0,0]
bas_appart=pd.DataFrame(appart.groupby(['cat_commune'])["nom_commune"].unique()).iloc[1,0]
haut_appart=pd.DataFrame(appart.groupby(['cat_commune'])["nom_commune"].unique()).iloc[2,0]
```

```python
bas_appart = airbnb["nom_commune"].isin(bas_appart)
moyen_appart = airbnb["nom_commune"].isin(moyen_appart)
haut_appart = airbnb["nom_commune"].isin(haut_appart)
```

```python
bas_appart = airbnb[bas_appart]
moyen_appart = airbnb[moyen_appart]
haut_appart = airbnb[haut_appart]
bas_appart.reset_index(drop=True, inplace=True)
moyen_appart.reset_index(drop=True, inplace=True)
haut_appart.reset_index(drop=True, inplace=True)
```

```python
bas_appart
```

```python
cat=[0,1,2]

list=[bas_appart,moyen_appart,haut_appart]
```

```python
for i in range(0,3):
    cat_com=[]
    for j in range(0,len(list[i])):
        cat_com.append(cat[i])
    list[i]['cat_commune']=cat_com
    
airbnb_cat=pd.concat([list[0],list[1],list[2]],ignore_index=True)  
```

```python
price_per_guest=[]
for i in range(0,len(airbnb_cat)):
    price_per_guest=airbnb_cat['prix']/airbnb_cat['numberOfGuests']
airbnb_cat['price_per_guest']=price_per_guest
```
